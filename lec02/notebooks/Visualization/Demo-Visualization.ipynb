```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme(style="whitegrid")
```

## 散点图案例

```python
dataset = pd.read_csv('lec04_datasets/Mall_Customers.csv')
dataset.head(10)
```

```python
X = dataset.iloc[:,[3,4]].values

plt.scatter(X[:, 0], X[:, 1], s = 10, c = 'black', label = 'Customer')
plt.title('Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
X_young = dataset[dataset['Age']<=30].iloc[:, [3,4]].values
X_handsome = dataset[(dataset['Age']>30)].iloc[:, [3,4]].values
plt.scatter(X_young[:, 0], X_young[:, 1], s = 10, c = 'red', label = 'Age<=30')
plt.scatter(X_handsome[:, 0], X_handsome[:, 1], s = 10, c = 'blue', label = 'Age>30')
plt.title('Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
company = pd.DataFrame(
    {'Company':['A','B','C','D','E'],
     'Country':['China', 'US', 'UK','China','US'], 
     'Industry':['Petro','Auto','Petro','Auto','Auto']}
)
company
```

## 掌握matplotlib

```python
#draw a line
fig = plt.figure(figsize=(8,8))
axes = plt.axes()
x = np.linspace(0,10,21)
print(x)
y = [v for v in x]
print(y)
axes.plot(x, y)
axes.set_xlabel('x');
axes.set_ylabel('y');
axes.set_title('This is a line.')
```

```python
#y = sin(x)
x = np.arange(0, 3 * np.pi, 0.1 )
y = np.sin(x)
z = np.zeros( 1 + int((3 * np.pi )/0.1))
plt.plot(x, y)
plt.plot(x, z)
```

```python
# draw sin(x), sin(x**2)
x = np.linspace(0, 10, 1000)
y = np.sin(x)
z = np.cos(x**2)

plt.figure(figsize=(8,4)) 

plt.plot(x,y,label="$sin(x)$",color="red",linewidth=2) 
plt.plot(x,z,"b--",label="$cos(x^2)$") 

plt.xlabel("Time(s)") 
plt.ylabel("Volt") 
plt.title("PyPlot Example") 
plt.legend()
```

```python
#different lines
t = np.linspace(0, 1, 21)
# red dashes, blue line, and green triangles
plt.plot(t, t, 'r--', label="$y=x$")
plt.plot(t, t**2, 'b-s', label="$y=x^2$")
plt.plot(t, t**3, 'g-*',label="$y=x^3$")
plt.plot(t, t**4, 'k-', linewidth=5, label="$y=x^4$")
plt.legend(loc='best')
```

```python
#特定方向的直线
plt.axvline(x=0.5, color='k', linestyle='--')
plt.axvline(x=0.25, color='r', linestyle='--')
plt.axhline(y=0.5, color='b', linestyle='--')
```

```python
#sub figure
x = np.linspace(1e-6, 1, 100) 
y = np.log(x)

fig, axes = plt.subplots(nrows=1, ncols=3)
colors = 'r b g k y'.split()
for i,ax in enumerate(axes):
    ax.plot(x, y, colors[i])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
max_row, max_col = axes.shape
print(max_row, max_col)

for i in range(max_row):
    for j in range(max_col):
        ax = axes[i][j]
        ax.plot(x, y, 'r')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_title('title')
    
fig.tight_layout()
```

## 通过实例掌握seaborn的绘图方法

```python
co2 = pd.read_csv('../../Datasets/co2-data/owid-co2-data.csv')
co2
```

```python
co2.isnull().any()
```

```python
co2 = co2.dropna(subset=['iso_code'])
co2 = co2[co2.iso_code != 'OWID_WRL']
co2
```

```python
df = co2[co2.year==2018]
df
```

```python
sns.histplot(data=df, x="co2_growth_prct", palette='muted')
#sns.histplot(data=df, x="co2_growth_prct", palette='muted', bins=[-50,0,50])
```

```python
sns.distplot(a=df['co2_growth_prct'], hist=True, kde=True, rug=True)
```

```python
sns.scatterplot(data=df, x="gdp", y="co2")
plt.xscale('log')
plt.yscale('log')
```

```python
df1 = df.sort_values('co2', ascending=False)
df1 = df1.head(15)
df1
sns.barplot(data=df1, x='iso_code', y='co2')
```

```python
df1 = df.sort_values('co2_per_capita', ascending=False)
df1 = df1.head(15)
sns.barplot(data=df1, x='iso_code', y='co2_per_capita')
```

```python
df = co2[co2.iso_code.isin(['CHN','USA','IND','RUS','JPN'])]
sns.lineplot(
    data=df,
    x="year", y="co2", hue="country", style="country",
    markers=True, dashes=True
)
```

```python
sns.lineplot(
    data=df,
    x="year", y="co2_per_capita", hue="country", style="country",
    markers=True, dashes=False
)
```

```python
df1 = df[(df.year>=1949) & (df.year<2021)]
sns.lmplot(
    x="year", y="co2",
    hue="country",
    col='country',
    data=df1)
```

```python
df1 = df[(df.year>=1949) & (df.year<2021)]
sns.lmplot(
    x="year", y="co2_per_capita",
    hue="country",
    col='country',
    data=df1)
```

```python
df1 = df[(df.year>=1949) & (df.year<2021)]
sns.lmplot(
    x="year", y="energy_per_gdp",
    hue="country",col='country',
    data=df1)
```
