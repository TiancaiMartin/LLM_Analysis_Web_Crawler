```python
#分词
def word_cut (texts):
    words_list = []
    word_generator = jieba.cut(texts, cut_all=False)  # 返回的是一个迭代器
    with open('hit_stopwords.txt',encoding = 'utf-8') as f:
        str_text = f.read()
    for word in word_generator:
        if word.strip() not in str_text:
            words_list.append(word)
            #print ('1')
    return ' '.join(words_list)

huolala_data['cut_comment'] = huolala_data['comment'].astype('str').apply(word_cut)
iran_data['cut_comment'] = iran_data['comment'].astype('str').apply(word_cut)

iran_data.head()
```

```python
#建立corpus
texts=[
    [word for word in document.split()]
    for document in data_shuffle['cut_comment']
]
#去除只出现一次的词
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token]+=1
texts=[
    [token for token in text if frequency[token]>1]
    for text in texts
]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
#tf-idf建模
#tf-idf表达
tfidf=models.TfidfModel(corpus)
corpus_tfidf=tfidf[corpus]
```

```python
# 找到最优的LDA主题数
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, random_state=100, 
                     update_every=1, chunksize=100, passes=30, alpha='auto', per_word_topics=True)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())
        print('number of topics = ',num_topics, " coherence = ", coherence_values[-1])

    return model_list, coherence_values
```

### 知乎伊朗头巾事件：按照时间顺序（日期）绘制河流图

为什么专门选择在知乎上的伊朗头巾事件呢？因为我们获得的数据中只有知乎的关于伊朗头巾问题的数据有连续一段时间的时间标签，可以用来观察一段时间内讨论主题的变化。尽管如此，我们也只有5天的数据。

#### 将数据中datetime标签中具体时分秒去掉

```python
zhihu_iran_data['date'] = zhihu_iran_data['datetime'].date()
```

```python
zhihu_iran_data['date'] = ''
for i in range(len(zhihu_iran_data)):
    zhihu_iran_data.iloc[i,2] = zhihu_iran_data.iloc[i,1].date()
zhihu_iran_data
```

将分词结果并入表中

```python
zhihu_iran_data['cut_comment'] = zhihu_iran_data['comment'].astype('str').apply(word_cut)
zhihu_iran_data
```

#### 开始对每条评论的主题权重求和

```python
zhihu_iran_data['Topic0']=''
zhihu_iran_data['Topic1']=''
zhihu_iran_data['Topic2']=''
zhihu_iran_data['Topic3']=''
```

使用语料库将分词后文本转为一串编码，随后使用上面最后训练的lda_model（对应使用了全部伊朗头巾事件的数据训练出的主题模型）对各个文本主题权重进行预测，并分别并入表中

```python
for i in range(len(zhihu_iran_data)):
    arr = np.array(zhihu_iran_data['cut_comment'][i].split(' '))
    bow = dictionary.doc2bow(arr)
    topics_lda = lda_model.get_document_topics(bow)
    for j in range(len(topics_lda)):
        z = topics_lda[j][0]
        zhihu_iran_data.iloc[i,z+4] = topics_lda[j][1]
```

数据处理

```python
#将空字符串转为0
data2 = zhihu_iran_data.replace(to_replace = '', value=0)
data2
```

```python
#将时间转为str后才能在画图事识别作为横坐标
data2['date'] = data2['date'].apply(lambda x:x.strftime('%Y-%m-%d'))
```

```python
df = data2.loc[:,('date','Topic0','Topic1','Topic2','Topic3')].groupby('date').sum()
```

```python
df
```

```python
import brewer2mpl

# plt.stackplot(df.index,df['doc_len'],df['Dominant_Topic'],baseline='sym')
# plt.show()

x = df.index
y1 = df['Topic0']
y2 = df['Topic1']
y3 = df['Topic2']
y4 = df['Topic3']

labels = ['topic0', 'topic1','topic2','topic3'] 

fig =plt.figure(figsize = (12,6))
ax = fig.add_subplot(111)

bmap = brewer2mpl.get_map('Set3', 'qualitative', 10)
colors = bmap.mpl_colors

ax.stackplot(x, y1, y2, y3, y4, labels=labels,colors=colors,baseline='sym') 
# ax.plot(x,y1,'bo-',markersize=22,alpha=0.2)
# ax.plot(x,y1+y2,'wo-',markersize=22,alpha=0.35)

ax.legend(loc=1,fontsize=12) #图例
ax.set_xticks(range(0,6,1))
ax.set_yticks(range(-200,200,25)) #y轴范围/刻度
ax.set_title('逐日主题显示流变图')
plt.savefig('逐日主题显示流变图.png')
plt.show()
```

最后没有什么惊人的结论，我们注意到随着时间的推移所有主题以相对比例比较均匀的方式变化，并没有在某一时刻出现某个主题的占比剧烈变化的情况。除此之外可以看出事件发生第二天热度达到高峰，随后便呈现下降趋势。
